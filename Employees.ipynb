```python
sns.boxplot(x = "Department", y = "Salary",data = df)
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"])
df["Year"] = df["JoinDate"].dt.year

years= sorted(df["Year"].unique())
averages = [df[df["Year"] == y]["Salary"].mean() for y in years]

plt.bar(years,averages)
plt.title("Averages Salary by joining date")
plt.xlabel("Years")
plt.ylabel("Average Salary")

plt.show()
```

```python
avg_years = df.groupby("Year")["Salary"].mean()
```

```python
plt.figure(figsize = (8,5))
plt.plot(avg_years, marker = "o", markerfacecolor = "Red", markeredgecolor = "Black")

plt.title("Average Salary by Years")
plt.xlabel("Year")
plt.ylabel("Average Salary")

plt.show()
```
